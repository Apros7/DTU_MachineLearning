```python
import pandas as pd
```

```python
data = pd.read_csv("degree_of_freedom.csv")
data.head()
```

```python
countries = data["Country/Territory"].to_list()
region = data["Region"].to_list()
status = data["Status"].to_list()

country_dict = {c: (r, s) for c, r, s in zip(countries, region, status)}
```

```python
life_expect = pd.read_csv("LifeExpectancyData.csv")
countries = life_expect["Country"].to_list()
countries_region = [country_dict[c][0] for c in countries]
countries_status = [country_dict[c][1] for c in countries]
df = pd.DataFrame()
df["Countries"] = countries
df["Region"] = countries_region
df["Status"] = countries_status
```
